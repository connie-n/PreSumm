```python
#1차
def samplefunc(df): 
    sample_list = []
    for i in range(len(df)):
      dialogue = df['dialogue'][i]
      summary = df['summary'][i]
      sample_list.append([dialogue, summary])
    return sample_list

#2차
def samplefunc(df): 
    sample_list = []
    for i in range(len(df)):
      dialogue = df['dialogue'][i]
      summary = df['summary'][i]
      #dialogue = dialogue.split(',')
      #summary = summary.split(',')
      sample_list.append({'src': dialogue,
                          '@tgt': summary})
    return sample_list

#3차
def samplefunc(df): 
    sample_list = []
    for i in range(len(df)):
      dialogue = df['dialogue'][i]
      dialogue += " "+'@highlight'+" "+df['summary'][i]

      #summary = df['summary'][i]
      #dialogue = dialogue.split(',')
      #summary = summary.split(',')
      sample_list.append(dialogue)
    return sample_list
```

```python
def samplefunc(df): 
    sample_list = []
    for i in range(len(df)):
      dialogue = df['dialogue'][i]
      dialogue+= '\r\n'
      dialogue+= '@highlight'+ '\r\n'+ " "+df['summary'][i]
      
      #summary = df['summary'][i]
      #dialogue = dialogue.split(',')
      #summary = summary.split(',')
      sample_list.append([dialogue])
    return sample_list
```

```python
#data split for train
for i in range(len(train_df)):
  json_list = samplefunc(train_df)[i]

  path = '/content/drive/MyDrive/Colab/github/PreSumm/SAMSum/split_data'


  file_name = os.path.join(path,"train.""{:d}.json".format(i))


  json_string = json.dumps(json_list, indent=4, ensure_ascii=False)
  with open(file_name, 'w') as json_file:
      json_file.write(json_string)

```

```python
#data split for test
for i in range(len(test_df)):
  json_list = samplefunc(test_df)[i]

  path = '/content/drive/MyDrive/Colab/github/PreSumm/SAMSum/split_data/test'


  file_name = os.path.join(path,"test.""{:d}.json".format(i))


  json_string = json.dumps(json_list, indent=4, ensure_ascii=False)
  with open(file_name, 'w') as json_file:
      json_file.write(json_string)
```

```python
#data split for valid
for i in range(len(val_df)):
  json_list = samplefunc(val_df)[i]

  path = '/content/drive/MyDrive/Colab/github/PreSumm/SAMSum/split_data/valid'


  file_name = os.path.join(path,"valid.""{:d}.json".format(i))


  json_string = json.dumps(json_list, indent=4, ensure_ascii=False)
  with open(file_name, 'w') as json_file:
      json_file.write(json_string)
```
